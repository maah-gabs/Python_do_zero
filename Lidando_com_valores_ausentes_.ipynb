### Tratando os dados 

Neste caso, o melhor a se fazer é analisar os dados apenas no periodo que se tem eles no banco. Pois, principamente em DataSets grandes, qualquer tipo de preenchimento pode enviesar os dados e alterar a interpretação do DataSet. 

Em outros casos, podemos considerar as seguintes hipoteses: 

*  **Excluir**

   * Se os dados ausentes se encontram em pequeno numero, se estão de forma aleatória e sua ausencia não carrega significado. A exclusão da linha pode ser uma opção. Porém, se no caso é necessário utilizar alguma parte dela, pode - se utilizar apenas uma parte selecionada. 

* **Preencher**

    * Preencher os dados com valores estatísticos, sendo: média, mediana, moda ou zeros; 
    * Quando a distribuição de dados é normal, a **média** é a mais útil a ser utilizada. Em dados que que a distribuição está mais enviesada a **mediana** consiste em uma solução mais robusta,pois, ela é capaz de ser mais sensivel a *outliers*;
    *Indetificar a entrada ausente com algum valor que indique isso pode ser mais informativo, quando a ausência representa valor. Por exemplo, em dados numéricos preencher com zero, e em categóricos criar uma categoria "Desconhecido". Atenção, pois os zeros não podem ser levados em consideração em análises estatísticas.
