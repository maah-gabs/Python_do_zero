## Exemplo 01: 

Neste exmplo, vamos utilizar os dados que estão disponibilizados no portal [Inside Airbnb](http://insideairbnb.com). 

Para esse notebook vamos utilizar dados que estão disponíveis no portal da cidade do Rio de Janeiro [aqui](https://data.insideairbnb.com/brazil/rj/rio-de-janeiro/2020-06-19/data/listings.csv.gz).

```
# Importando o Pandas
import pandas as pd

# Para não limitar a visualização do número de colunas 
pd.set_option('display.max_columns', None)
```

```
# Importando os dados 
df = pd.read_csv('http://data.insideairbnb.com/brazil/rj/rio-de-janeiro/2020-06-19/data/listings.csv.gz')
```

```
# Visualizar as cinco primeiras entradas 
df.head()
```

Ao realizar uma análise inicial nas primeiras, já percebemos que a coluna `'price'` está em formato `string`, desta forma, será necessário transformar em formato numérico para que a couna possa ser utilizada. 

```
# Verifcando as entradas da coluna price
df['price']
```

Perceba, que em consequencia do formato do texto **não** será possível converter diretamente o valor acima em float. Para isso, deveremos tratar a coluna inicialmente. 

Para compreendermos melhor o processo, vamos manipular apenas uma variável e associar ela a uma String bem parecida com a que encontrams no nosso DataFrame.

```
# Associando o valor da variável 
valor = '$244.00'
```

Sabemos que para iniciar o processo de transformação da `String` em `Float`, precisamos tirar o $. Neste caso, utilizaremos o método `.replace ( )`.

```
# Limpando a variável 
float(valor.replace("$",''))
```

## Funções Lambda 

As funções lambdas, consistem naquelas que apresentam funções um pouco mais restritivas em Python. 
Quando a função for ser utilizada várias vezes no seu código, talvez seja melhor utilizar funções maiores, porém, como o intuito desta aula é fazer limpeza e a manipuação do DataSet, podemos utilizar uma função mais simples como a `lambda`. 

```
# Exemplo de Função 
def elevar_quadrado(numero):
  resposta = numero ** 2 
  return resposta
```

```
elevar_quadrado(5)
```

```
# Exemplo de Função 

def somar_numeros (a,b):
  resultado = a + b
  return resultado
```

```
somar_numeros(3,5)
```

```
# "$244.00" ==> 244.00
def converter_str(minha_string):
  resposta = float(minha_string.replace("$", ""))
  return resposta
```

```
converter_str("$244.00")
```

```
# Criando uma função lambda
lambda x: float(x.replace("$",""))
```

Com a função pronta, vamos aplicar na coluna que queremos alterar. 

```
df["price"].apply(converter_str)
```

```
# aplicando a função à coluna
df["price"].apply(lambda x: float(x.replace("$", "")))
```

Perceba acima, que mesmo ao tentar executar as duas funções, ela retorna o mesmo erro demonstrado acima, será que a gente consegue definir o que é? 

Veja, que ao especificar o erro o Python diz: 
`ValueError: could not convert string to float: '4,161.00'`, sabemos que o Python reconhece apenas ponto, então, no decorrer do nosso processo, acabamos perdendo a noção e "assumindo" que as estruturas de todo o dataset seriam apenas de "$", ou seja, o caractere que atrapalharia a conversão em float. Porém, o erro nos alertou para outra alteração que precisa ser feito. Sendo assim, temos que fazer uma outra limpeza, que será demonstrada na linha abaixo. 

```
# Associanoo uma nova função a variável 
valor = "$4,161.00"

# Acrescentando um passo a função 
float(valor.replace(",", "").replace("$", ""))
```

Agora, vamos aplicar a função a partir do que descobrimos acima. 

```
df['price'].apply(lambda x:float(valor.replace(",", "").replace("$", "")))
```

## Exemplo 02 

Como segundo exemplo, vamos utilizar dados disponibilizados pelo [Kaggle](https://www.kaggle.com/ravels1991/brasil-solicitaes-de-refugiado?) sobre a situação dos refugiados no Brasil.

Os dados reais podem ser encontrados [aqui](https://www.kaggle.com/ravels1991/brasil-solicitaes-de-refugiado?).

```
# carregando dados 
df = pd.read_excel("https://www.dropbox.com/s/69pajozeme6ezjc/solicitacoes-de-reconhecimento-da-condicao-de-refugiado.xlsx?dl=1")

# Visualizar as 5 primeiros entradas 
df.head()
```

Quando comparado com o outro conjunto de dados que utilizamos esse é bem mais simples, certo? Porém, ao plotar as cinco primeiras entradas acima, você perdebe que as Strings estão todas em letras maiúsculas e gostariamos de modifica - la. 

```
# Verificando as entradas 
df['Tipo de Alertas e Restrições']
```

```
# Aplicando o capitalize
df['Tipo de Alertas e Restrições'].apply(lambda x: x.capitalize())
```

```
df
```

Perceba que ao chamar o DataFrame na execução anterior ainda não foi alterado, desta forma, precisamos armazenar essas alterações em uma variável, ação que faremos a seguir: 

```
# Aplicando a função relacionada a coluna
df['Tipo de Alertas e Restrições']= df['Tipo de Alertas e Restrições'].apply(lambda x: x.capitalize())
```

```
df.head()
```

Agora o DataFrame está com o formato desejado. 

Agora, vamos fazer o mesmo com a coluna `Nacionalidade`.

```
# Aplicando a função 
df['Nacionalidade'] = df['Nacionalidade'].apply(lambda x: x.title())
```

```
# Verificando a alteração do DataSet 
df.head()
```

## Exemplo 03 

Para o último exemplo, vamos utilizar os dados relacionados a taxa de homicídio no Brasil em 2017. E os dados estão disponíveis no [Kaggle](https://www.kaggle.com/diegomariano/taxa-de-homicdios-no-brasil-no-ano-de-2017).

Neste exemplo, encontraremos uma outra dificuldade que é muito comum e que é necessário muita atenção. 

**OBS:** É muito importante conhecermos os dados muito bem. 

```
# importando os dados
df = pd.read_csv("https://www.dropbox.com/s/16nz0yikhma091u/homicidios_brasil_2017.csv?dl=1")
```

```
df.head()
```

Apesar da importação ter sido feita da mesma forma como sempre fazemos, ao visualizar o nosso df, percebemos que ele está todo estranho e esse formato impede que trabalhemos com ele. 

Isso pode acontecer, porque alguns arquivos `csv` *(Comma Separated Values)* tem como separador padrão a vírgula "," e o arquivo que foi importado está utilizando o ponto e virgula como separador ( ; ). 

Esse é um problema muito simples e que pode ser resolvido de uma forma bem simples. 

```
# carregando o arquivo com o separador correto
df = pd.read_csv("https://www.dropbox.com/s/16nz0yikhma091u/homicidios_brasil_2017.csv?dl=1", sep=";")

df
```

Ao executar a linha acima e dar uma olhada rápida no conjunto de dados, percebemos que a Sigla de São Paulo e Goias estão escritas erradas. 

Precisamos consertar esse detalhe para que a gente possa fazer uma boa análise de dados. 

Para isso, vamos criar um dicionário que receberá como chave a entrada incorreta e como valor a correção. 

Depois a gente utilizará o método `.replace ( )` para que a correção seja aplicada. 

```
# Criando o dicionário para correção 
correcao = {"G": "GO", "São Paulo": "SP"}

# Aplicando a correção 
df['sigla'] = df['sigla'].replace(correcao)

# Verificando o resultado
df
```
