```
# Carregando uma planilha 
from openpyxl import load_workbook

workbook = load_workbook(filename = 'controle_da_empresa.xlsx')

# Verificando as abas 
sheet = workbook.active
```

Neste momento, vamos estar sempre modificando as entradas de nossa planilha e necessitamos conferir os resultados com determinada frequencia. Desta forma, vamos utilizar o mesmo código que criamos na aula anterior para que sejamos capazes de facilitar as nossas entradas. 

```
# criando função para leitura dos dados
def print_rows():
  for row in sheet.iter_rows(values_only=True):
    print(row)
```

```
# testando a função
print_rows()
```

## Adicionando colunas 

Para adicionar colunas, vamos utilizar o método `insert_cols`, com o índice em que será inserida como parametro. LEMBRE - SE a indexação do Excel se inicia com 1. 

```
# Adicionando uma coluna 
sheet.insert_cols(idx=1)
print_rows()
```

Também é possivel criar uma coluna mais de uma vez, passando assim pelo parametro `amount` com a quantidade de colunas que desejamos criar. 

```
# adicionando uma coluna a partir de um índice 
sheet.insert_cols(idx=3, amount=3)
print_rows()
```

## Deletando colunas 
Também é possivel deletar as colunas a partir do comando de `delete_cols`, os parametros passados anteriormente. 

Caso você tenha criado novas colunas, como assim nós fizemos aqui. É importante que estejamos atentos ao estado **atual** dos índices. 

Sendo assim, vamos excluir as colunas extras que foram criadas. 

```
# Deletando as colunas 
sheet.delete_cols(idx=3, amount=3)
print_rows()
```

```
# Deletando colunas 
sheet.delete_cols(idx=1)
print_rows()
```

## Manipulando Linhas 
Da mesma maneira que utilizamos `insert` e `delete` cols para que possamos manipular colunas, utilizaremos o mesmo tipo de método para as linhas. Desta forma, ao invés de usar col utilizaremos rows. 

A indexação e o princípio das funções continuem os mesmos. 

```
# Criando linhas 
sheet.insert_rows(idx=1)
print_rows()
```

```
# inserindo mais de uma linha de uma só vez 
sheet.insert_rows(idx=5, amount=5)
print_rows()
```

```
# deletando colunas 
sheet.delete_rows(idx=1)
print_rows()
```

```
# deletando mais de uma linha de uma vez
sheet.delete_rows(idx=5, amount=5)
print_rows()
```
