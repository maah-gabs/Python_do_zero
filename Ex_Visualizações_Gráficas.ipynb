```
#Importar as bibliotecas necessárias
import pandas as pd 
import matplotlib.pyplot as plt 
from google.colab import drive
```

```
#Montar o Drive 
drive.mount('/content/gdrive')
```

```
#Importar a planilha 
planilha = pd.read_excel("/content/gdrive/My Drive/controle_da_empresa.xlsx", None)

#associar a planilha vendas em df
df = planilha['Vendas']

#ver as cinco primeiras entradas
df.head()
```

```
df['Item'].value_counts()
```

```
df['Vendedor'].value_counts().plot.bar();
```

```
#plotar um histograma com o total de vendas

plt.hist('Total de Vendas', data=df)
plt.title("Total de Vendas")
plt.ylabel("Quantidade")
plt.xlabel("Valor")

plt.show()
```

```
# rode a célula para transformar a coluna Datas para o formato datetime
df['Data da Venda'] = pd.to_datetime(df['Data da Venda'], infer_datetime_format=True)

#Plotar um gráfico de linha de vendas ao longo do tempo
plt.plot('Data da Venda','Total de Vendas', data=df)
plt.title("Total de Vendas")
plt.ylabel("Total de Vendas")
plt.xlabel("Data da Venda")

plt.show()
```

```
df.groupby('Vendedor')['Total de Vendas'].sum().plot.bar();
```

```
## Plotar um gráfico de Waffle 
a = dict(df['Item'].value_counts())

labels = a.keys()
valores = a.values()

plt.pie(x=valores, labels=labels)
plt.title("Item mais vendido")

plt.show()
```

```
fig = plt.figure (
                  FigureClass = Waffle, 
                  rows = 4,
                  columns = 10,
                  values = a, icons = 'dot-circle', icon_size = 12, 
                  legend = {'loc': 'upper left', 'bbox_to_anchor':(1,1)}, 
                  icon_legend = True
                  )
fig.set_tight_layout(False)
plt.show();
```
