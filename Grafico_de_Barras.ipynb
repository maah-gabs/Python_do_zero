# Módulo 03: Aula 04 - Gráfico de Barras 

O gráfico de barras consiste em uma importante ferramenta para variáveis discretas. Apesar de ser visualmente parecido, com histograma e por conta disso, são muitas vezes confundidos. Porém, veremos melhor sobre esse gráfico na aula. 

Esse gráfico é composto por barras em formato retangular. O comprimento ou altura do gráfico é proporcional ao valor que ela representa. 



## DataSet utilizado nesta aula 

Para essa aula, vamos utilizar um DataSet disponivel no Portal da Governança Corporativa e Transparencia que está descrito a gestão financeira, novos projetos em gestação e qual o andamento das obras em evoluções. O dataset está disponível [neste link.](https://transparencia.metrosp.com.br/). O arquivo em `csv` está disponível [aqui](https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/entrada_passageiro_linha1.csv).

```
# Importar as bibliotecas 
import pandas as pd 
import matplotlib.pyplot as plt
```

```
# Importar os dados do DataFrame 
df = pd.read_csv("https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/entrada_passageiro_linha1.csv")

#Ver as cinco primeiras entradas
df.head()
```

```
# Ver as cinco últimas entradas 
df.tail()
```

Perceba, que ao plotar o DataSet, temos uma linha com o TOTAL. Porém, essa informação no DataFrame vai atrapalhas os nossos dados, sendo assim, vamos ELIMINAR essa linha. 



```
#df.drop(index = 23, axis = 0, inplace = True)
```

```
# Plotar o Gráfico de Barras 

df.plot(x = "Estação", y = "Entradas", kind="bar")

plt.title("Entrada Média de Passageiros - (dez/18)")
plt.xlabel("Estação")
plt.ylabel("Entradas")
plt.show()
```
