# Módulo 03: Aula 02 - Conhecendo o Matplotlib 

Consiste na principal biblioteca para visualização de Python. A arquitetura do Matplotlib é composta por três camadas: 

1. **Scripting Lyer**
2. **Artist Layer**
3. **Backend Layer**

Para ter uma compreensão de como o pacote funciona, é bem importante compreender que a biblioteca foi criada para renderizarem objetos do tipo *figure*

```
# Importar os pacotes 
import numpy as np
import matplotlib.pyplot as plt
```

```
# Gerar valores demonstrativos
np.random.seed (42)
x = np.arange(10)
y = np.random.normal(size =10)
```

```
# Plotar os gráficos 
plt.plot(x,y)

#para definir o título
plt.title("Exemplo de Título")

#definir o label do X e do Y
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")

plt.show()
```

## Abordagem Interface do Pyplot. 

Como exemplificado, anteiormente, existe uma complexidade muito importante por trás da biblioteca e artificios podem ser utilizadas a navegar pelo mais baixo nível ao mais complexo. Na abordagem orientada a objetos como é o caso do Matplotlib é necessário que as `figures` e `axes` devem ser criadas explicitamente. 


```
# Criando números a serem plotados 
x = np.arange(10)

#Plotando um gráfico de linhas. 
plt.plot(x,x);
```

No exemplo abaixo, colocaremos título no nosso gráfico, e assim, seguiremos a mesma abordagem. 

```
# Criando o plot 
plt.plot (x,x)

#Definindo o título 
plt.title("Linear")

#definindo os titles para os eixos 
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")

plt.show()
```

De maneira similar, se eu quiser, acrescentar objetos ao `axes`, a gente adiciona apenas adicionando novas funções sequencialmente. 

```
# criando o plot 
plt.plot (x, x, label = "Linear")

#adicionando ao plot 
plt.plot (x, x**2, label = "Quadrado")

#defindo os títulos 
plt.title ("Abordagem Pyplot")
plt.ylabel ("Valores de Y")
plt.xlabel("Valores de X")

#Plotando a legenda 
plt.legend()

#mostrando o plot 
plt.show()
```
